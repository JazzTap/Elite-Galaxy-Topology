```python
import json
import gudhi
with open("./elite-galaxy-00.json") as f:
    systems = json.load(f)
```

```python
systems[5]
```

```python
print("RipsComplex creation from points")
rips = gudhi.RipsComplex(points=[[u['location'][0], u['location'][1]/2] for u in systems], max_edge_length=50)
simplex_tree = rips.create_simplex_tree(max_dimension=2)
```

```python
result_str = 'Rips complex is of dimension ' + repr(simplex_tree.dimension()) + ' - ' + \
    repr(simplex_tree.num_simplices()) + ' simplices - ' + \
    repr(simplex_tree.num_vertices()) + ' vertices.'
print(result_str)
```

```python
diag = simplex_tree.persistence() # homology_coeff_field=2, min_persistence=0.01)
gudhi.plot_persistence_barcode(diag)
```

```python
simplex_tree.persistence_intervals_in_dimension(1)
```

```python
filtration = [u for u in simplex_tree.get_filtration()]
```

```python
fmt = '%s -> %.2f'
for filtered_value in simplex_tree.get_filtration():
    print(fmt % tuple(filtered_value))
```
